````python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams.update({'font.size': 20, 'figure.figsize': (10, 8)})

# Load the dataset
df = pd.read_csv("travel-times.csv")

# Print the shape of the dataset
print("Shape of the dataset:", df.shape)

# Print the first 15 rows of the dataset
print("First 15 rows of the dataset:")
print(df.head(15))

# Get information for the features with missing values
print("Information for features with missing values:")
print(df.info())

# Drop duplicate values by keeping only the last instance
df.drop_duplicates(keep='last', inplace=True)

# Calculate the total number of missing values on each column
print("Total number of missing values on each column:")
print(df.isnull().sum())

# Create two copies of the dataframe
df_copy1 = df.copy()
df_copy2 = df.copy()

# Drop rows with missing values from the 1st copy
df_copy1.dropna(inplace=True)

# Drop columns with missing values from the 2nd copy
df_copy2.dropna(axis=1, inplace=True)

# Get summary statistics and see the correlation between the numerical columns
print("Summary statistics:")
print(df.describe())

# Remove non-numeric columns
numeric_columns = df.select_dtypes(include=[np.number]).columns
df_numeric = df[numeric_columns]

print("Correlation between numerical columns:")
print(df_numeric.corr())

# Show rows 11 to 14
print("Rows 11 to 14:")
print(df.iloc[10:14])

# Create a subset with trips occurred on November 23, 2011 and January 6, 2012
subset = df[(df['Date'] == '2011-11-23') | (df['Date'] == '2012-01-06')]
print("Subset of trips on November 23, 2011 and January 6, 2012:")
print(subset)

# Produce a scatterplot between Distance and TotalTime
plt.figure()
plt.scatter(df['Distance'], df['TotalTime'])
plt.xlabel('Distance')
plt.ylabel('TotalTime')
plt.title('Scatterplot: Distance vs TotalTime')
plt.show()

# Produce a boxplot for AvgSpeed
plt.figure()
plt.boxplot(df['AvgSpeed'])
plt.ylabel('AvgSpeed')
plt.title('Boxplot: AvgSpeed')
plt.show()

# Produce a boxplot for AvgMovingSpeed
plt.figure()
plt.boxplot(df['AvgMovingSpeed'])
plt.ylabel('AvgMovingSpeed')
plt.title('Boxplot: AvgMovingSpeed')
plt.show()
````
